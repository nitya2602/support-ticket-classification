# Sentiment Analysis with FEW stopwords #

```python

import string
import pandas as pd

from collections import Counter

from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.stem import WordNetLemmatizer

from nltk.tokenize import word_tokenize
import xlrd 
def listToString(s):   
    str1 = ""  
    for ele in s:  
        str1 += ele 
        str1+=" "
    return str1

def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    if score['neg'] > score['pos']:
        print("NEGATIVE SENTIMENT")
        return "NEGATIVE SENTIMENT"
    elif score['neg'] < score['pos']:
        print("POSITIVE SENTIMENT")
        return "POSITIVE SENTIMENT"
    else:
        print("NEUTRAL SENTIMENT")
        return "NEUTRAL SENTIMENT"

        
dataframe=pd.read_excel('TicketData2.xlsx')
x1=dataframe["CustomerReportedCommunication"]
a1=x1[0:100]
stopwords1=['good', 'morning','afternoon','evening','thanks','hello','regards','thank','kind regards',',']  
loc = ("TicketData2.xlsx") 
  
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0) 
a=[]
for i in range(sheet.nrows-1):
      print(i+1,end=".")
      print(a1[i],"  - ",end=" ")  
      l1=[]
      l1=sheet.cell_value(i+1,6)
      l2=l1.lower()
      tokenized_words = word_tokenize(l2, "english")
      final_words = []
      for word in tokenized_words:
         if word not in stopwords1:
            final_words.append(word)
      a.append(str(sentiment_analyse(listToString(final_words))))
      print("\n\n\n\n")
w = Counter(list(a))
print(w)


temp = dict()
temp['FEW STOPWORDS'] = w
emotion_count_comparision_list.append(temp)
```
